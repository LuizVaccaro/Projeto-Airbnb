# Panorama do COVID-19 no Mundo e na França

COVID-19 é uma doença infecciosa causada por conta do coronavírus.
 
Ainda sabemos muito pouco deste novo vírus, mas após quase 3 anos de pandemia, já temos algumas certezas, ele é transmitido principalmente por meio de gotículas provenientes de tosses ou espirros de pessoas infectadas, a gravidade dos sintomas varia muito de pessoa para pessoa.
 
<center><img alt="Colaboratory logo" width="50%" src="https://images.pexels.com/photos/4031867/pexels-photo-4031867.jpeg?auto=compress&cs=tinysrgb&w=1260&h=750&dpr=1"></center>
 
Observamos que os governos ao redor do mundo tentaram diversas alternativas para frear a dissipação do vírus em seus países, muitas podem se considerar com sucesso e outras sem sucesso. 
 
Sabemos após esses quase 3 anos do vírus que apesar de toda a tecnologia atual não estávamos preparados para uma doença tão silenciosa, mortal e capaz evoluir de uma forma tão rápida. 
 
Então nesta análise vamos observar um pouco mais de como está a situação da pandemia em torno do mundo e especificamente no país da França. 
 
Por quê França? Porque após uma inicial análise da situação da covid-19 no mundo, observei que a França é um dos países com o maior número de casos no mundo, mas não é um dos maiores países com número de mortos.
 
Vamos tentar entender como eles conseguiram apesar de uma explosão de casos conseguir manter o índice de mortalidade abaixo dos outros países que tiveram muitos casos e também muitas mortes, como o Brasil e Estados Unidos. 

## Obtenção dos Dados

### Dicionário de Variáveis

* new_testes_per_thousand - novos testes por mil pessoas

* new_testes_smoothed_per_thousand - suavização dos dados dos novos testes por mil pessoas

* tests_per_case - testes por caso

#Análise sobre o número de casos e mortos

Vamos observar os gráficos de barras do número de casos e mortos dos 5 primeiros países em cada situação.

```
#Selecionando as colunas utilizadas no gráfico.
top5_td = dfx.loc[dfx.date == '2023-01-05', ['location', 'total_deaths']].sort_values(by="total_deaths", ascending=False) [0:5]
top5_tc = dfx.loc[dfx.date == '2023-01-05', ['location', 'total_cases']].sort_values(by="total_cases", ascending=False) [0:5]

#Plotando o gráfico
fig, (ax1,ax2) = plt.subplots(nrows=2, ncols=1, figsize=(10,8))

ax1.barh('location', 'total_cases', data=top5_tc, color = ['blue','#f4c430', '#191970','#111', '#ffd700'])
ax1.set_title("Países Com Maior Número de Casos Por Covid 19")
ax1.set_xlabel("Número de Casos - A cada 200 mil")
ax1.set_ylabel("País")

ax2.barh('location', 'total_deaths', data=top5_td, color = ['blue','#ffd700', '#f4c430','red', 'green'])
ax2.set_title("Países Com Maior Número de Mortes Por Covid 19")
ax2.set_xlabel("Número de Mortes - A cada 200 mil")
ax2.set_ylabel("País")


plt.tight_layout()
```

Aqui podemos observar por este gráfico de barras os 5 países com o maior número de mortos no mundo. Vemos como o Estados Unidos foi extremamente afetado pelo número de casos e mortos. Sendo líder em ambos os índices, uma diferença de quase 400 mil mortos para o segundo lugar Brasil. Outra situação que podemos notar, contradizendo aos Estados Unidos, é que por mais que tenhamos um número maior de casos, não necessariamente representa o número de mortos, como no caso de Alemanha e França.

* Temos Estados Unidos, Índia, França, Alemanha e Brasil com o maior número de casos.
* Temos Estados Unidos, Brasil, Índia, Rússia e México com o maior número de mortos.

México por exemplo não está entre os 10 países com o maior número de casos mas está entre os 5 com o maior número de mortos. Temos que levar em consideração por ser um país subdesenvolvido tendo um menor preparo na área da saúde.
Também pode representar uma subnotificação do número de casos ou até mesmo demora para atualização dos dados, temos muitas variáveis que incidem na qualidade dos nossos dados.


#### Gráfico da evolução no número de mortes por Covid-19

```
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10,8))

df.loc[df.location == "World", "total_deaths"].reset_index(drop=True).plot(kind="line", ax=ax)
ax.set_title("Total de Mortes")
ax.set_xlabel ("Número de Dias")
ax.set_ylabel ("Número de Casos")

plt.tight_layout()
```

Podemos observar a evolução no número de mortos, como nos primeiros meses tivemos uma explosão no número de mortes e após os 800 dias, mais de dois anos, vemos o início de platô na linha de crescimento nas mortes.

Representado principalmente pelo aumento no número de vacinados ao redor do mundo e pela criação de anticorpos em relação às variantes. Fazendo com que o vírus tenha uma mutação mais devagar e mais fraca.

Abaixo com o gráfico da comparação de novos casos com o total de mortes, podemos entender como foi o crescimento do índice de mortalidade com o aumento dos casos, observar de uma forma mais clara como foi a escalada do número de mortes com o grande número de novos casos e sobrecarregamento os sistemas de saúde ao redor do mundo.

E observar de forma concreta que após os 800 dias que mesmo com um aumento no número de casos temos uma estabilidade na taxa de mortalidade. Fortalecendo as razões que pontuamos logo acima, sobre a importância das vacinas e dos anticorpos na batalha contra o vírus.


```
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10,8))

#total de casos
df.loc[df.location == "World", "new_cases"].reset_index(drop=True).plot(kind="line", ax=ax)
ax.set_title("Total de Casos")
ax.set_xlabel ("Número de Dias")
ax.set_ylabel ("Número de Novos Casos e Número de Mortes")

#total de Mortes
df.loc[df.location == "World", "total_deaths"].reset_index(drop=True).plot(kind="line", ax=ax, color="red")


plt.tight_layout()
```

####Agora iremos ver um gráfico de dispersão 

Iremos observar nesse gráfico como o desenvolvimento do país impactou no número de mortes. Com os dados da renda per capita pelo número de mortos teremos uma ideia de como impacta os países. 

```
#Gráfico de dispersão 
dfd = dfx.loc[dfx.date == '2023-01-05', ['gdp_per_capita', 'total_deaths']].sort_values(by="total_deaths", ascending=False)

fig, ax = plt.subplots(figsize=(8,6))
dfd.plot.scatter(x='gdp_per_capita', y='total_deaths', ax=ax, s=50,)
ax.set_title("Dispersão dos Dados")
ax.set_xlabel("Renda Per Capita")
ax.set_ylabel("Total de Mortes")
plt.show()
```

Conforme ressaltado durante a análise dos 5 países com o maior número de mortos no mundo, retirando os Estados Unidos, grande parte dos países estão na categoria de subdesenvolvidos. Onde normalmente temos um investimento menor na área da saúde e com menos acesso a cuidados médicos.

Podemos ver que até 40 mil dólares per capita temos a maioria dos maiores índices de mortalidade.


## Análise Exploratória para a França

A escolha pelo país da França para uma análise mais aprofundada é motivada pelo seu alto número de casos, sendo o terceiro país com o maior número, e o seu índice menor de mortalidade, sendo o nono país no número de mortes. 
 
Vamos analisar se os números podem nos dar algumas respostas em relação a estas perguntas:
 
* Quando foi a primeiro primeiro caso e primeira morte por Covid-19 na França.
 
* Por quê França teve um alto índice de casos e não tão alto número de mortos.
 
* Cenário atual da doença no país. 
 
Vamos entender por meio dos dados a forma como a frança foi impactada pelo covid-19 e algumas atitudes que levam a ela ter conseguido diminuir o seu índice de mortalidade apesar do alto número de casos. 



<center><img src="https://images.pexels.com/photos/5781917/pexels-photo-5781917.jpeg?auto=compress&cs=tinysrgb&w=1260&h=750&dpr=1" height="300px"></center>

Primeiramente vamos observar os primeiros e os últimos casos relatados pela França e começar a entender como a doença começou a se desenvolver no país e como está no dia de hoje. 

```
df_france = df.loc[df.location == "France"].copy()
df_france = df_france.fillna(0)

# 5 primeiras entradas do país
df_france.head()
```

```
# 10 últimas entradas do país
df_france.tail(10)
```

```
#Identificação Primeiro Caso de Covid-19 na França 
df_france.loc[df_france['new_cases']!=0.0].head(1)
```

Para entender a forma como iniciou a transmissão da Covid-19 no país veremos a data do primeiro caso que foi identificado no dia 21/01/2020. Para levarmos em consideração, o primeiro caso confirmado no mundo foi no dia 17/11/2019 na China.


```
#Identificação da primeira morte na França
df_france.loc[df_france['new_deaths'] != 0.0].head(1)
```

A primeira morte decorrida da Covid-19 na França foi identificada no dia 15/02.

```
from datetime import date
delta = df_france.date.loc[78045] - df_france.date.loc[78023]
print('Passaram-se {} dias entre o primeiro caso e a primeira morte na França.'.format(delta.days))
```

```
#Gráficos de linhas nas escalas linear e logarítmica para casos totais e mortes totais.
x = list(range(1085))
f = df_france['total_cases']
f2 = df_france['total_deaths']

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(12,9))
ax1.set_title('Casos na França')
ax1.set_xlabel('Dias')
ax1.set_ylabel('Casos')
ax1.plot(x, f)

ax2.set_title('Casos França (Escala Logarítmica)')
ax2.set_xlabel('Dias')
ax2.set_ylabel('Casos')
ax2.set_yscale('log')
ax2.plot(x,f)

ax3.set_title('Mortes França')
ax3.set_xlabel('Dias')
ax3.set_ylabel('Mortes')
ax3.plot(x,f2)

ax4.set_title('Mortes na França (Escala Logarítmica)')
ax4.set_xlabel('Dias')
ax4.set_ylabel('Mortes')
ax4.set_yscale('log')
ax4.plot(x,f2)
plt.tight_layout()
```

Agora podemos observar de uma forma mais clara como a doença se desenvolveu na França, o país que chegou a ser um dos epicentros da pandemia.
 
Se observarmos a coluna da esquerda onde estão os gráficos com escala linear podemos ver como a pandemia iniciou no país com uma curva acentuada de casos e mortes, durante os dias 400 até 600 teve uma diminuição no ritmo dos casos, entretanto o número de mortes não acompanhou o ritmo dos casos, provavelmente por um sistema de saúde ainda sobrecarregado por conta da primeira onda do vírus. 
 
Após os 600 dias vemos mais uma explosão no número de casos, poderemos ver de forma mais concreta no gráfico abaixo com a comparação de novos casos e número total de mortos, onde mostrará como foi o aumento em relação às outras ondas. 

A segunda onda em relação à primeira, vemos como foi menos impactante no número de mortos, por conta das vacinas e do melhor preparo dos profissionais em combater a doença. 
 
Na escala logarítmica, vemos a expansão da doença de uma forma diferente, nota-se como nos primeiros anos ela teve um crescimento de casos e mortes avassalador no país e com um crescimento constante nos números de casos mas uma estabilização no número de mortos. 
 
São duas análises que se completam no panorama de observação do crescimento da doença, de como a França apesar de um início avassalador com suas atitudes conseguiu frear a taxa de mortalidade. 


```
fig, ax = plt.subplots (figsize=(10,8))

#total de casos
df_france.loc[df_france.location == "France", "new_cases"].reset_index(drop=True).plot(kind="line", ax=ax, label='Teste')
#total de Mortes
df_france.loc[df_france.location == "France", "total_deaths"].reset_index(drop=True).plot(kind="line", ax=ax,color="red", label='2')

ax.set_title("Total de Casos")
ax.set_xlabel ("Número de Dias")
ax.set_ylabel ("Número de Casos")




plt.tight_layout()
```

Neste gráfico de novos casos e número de mortes conseguimos ver claramente a explosão dos casos após os 600 dias, mostrando de uma forma mais clara como a essa onda de casos teve impacto no país, e se formos comparar com o mesmo gráfico nos dados do mundo podemos ver o mesmo cenário no mundo, mostra como a doença impactou o planeta de uma forma muito parecida, apesar de medidas diferentes para diminuir o seu impacto. 

Também podemos o observar que após os 600 dias que apesar do aumento exponencial do número de casos no número de mortes houve um aumento menor em relação a primeira onda durante os dias 200 e 400, onde o país foi muito mais impactado pelo índice de mortalidade. 

####Vamos observar como se comportaram as internações nos leitos de hospitais e tratamento intensivo (ICU) durante a expansão do vírus no país.

```
fig, ax = plt.subplots (figsize=(10,8))

df_france.loc[df_france.location == "France", "hosp_patients"].reset_index(drop=True).plot(kind="line", ax=ax)

ax.set_title("Total de Pessoas Hospitalizadas")
ax.set_xlabel ("Dias")
ax.set_ylabel ("Número de Pessoas")
```

Podemos ver no gráfico acima e no abaixo como foram as hospitalizações e internações na terapia intensiva dos pacientes da Covid-19, eles mostram a forma como os hospitais foram impactados pela alta demanda de pacientes com casos mais agravados da doença. Podemos ver comparando com o gráfico dos casos e novos casos como teve um aumento considerável no número de hospitalizações quando aumenta o número de casos. 



```
fig, ax = plt.subplots (figsize=(10,8))

df_france.loc[df_france.location == "France", "icu_patients"].reset_index(drop=True).plot(kind="line", ax=ax)

ax.set_title("Total de Pessoas No Tratamento intensivo")
ax.set_xlabel ("Dias")
ax.set_ylabel ("Número de Pessoas")
```

No gráfico de internações no tratamento intensivo, vemos como diminuiu o número de pacientes com quadros graves da doença, influenciando diretamente no número de mortes. 

Apesar de ainda ter um volume considerável de pessoas indo aos hospitais com quadros mais complicados, não representou casos graves e necessidade de tratamento intensivos. 

Podemos ver que a diminuição de casos graves e número de mortes no gráfico abaixo onde demonstrará o número de vacinados e como coincide diretamente com essa diminuição de gravidade. 

```
fig, ax = plt.subplots (nrows=1, ncols=1, figsize=(10,8))

#total de casos
df_france.loc[df_france.location == "France", "people_vaccinated"].reset_index(drop=True).plot(kind="line", ax=ax)
ax.set_title("Total de Vacinados")
ax.set_xlabel ("Número de Dias")
ax.set_ylabel ("Pessoas Vacinadas")

plt.tight_layout()
```

Podemos observar que o maior número de vacinados foi entre os 400 a 600 dias que antecede a onda que vem após os 600 dias, mostrando como a vacinação contribuiu de uma forma positiva para a diminuição nos casos mais graves da doença, diminuindo o sobrecarregamento dos hospitais e consequentemente o índice de mortalidade. 

Além de fatores como restrições, lockdown e outras atitudes por parte do governo que podem também ter contribuído para a diminuição, mas como não temos dados para compararmos fica mais complicado saber até que ponto beneficiou.

Obs: por conta possivelmente algum erro no dataset o qual não somou o números de vacinados temos uma queda brusca no gráfico de vacinados.

Podemos ver como as vacinas e os anticorpos foram extremamentes importantes para a diminuição nos índices de mortalidade ao redor do mundo. Conforme ressaltado acima, ainda é difícil concluir se as estratégias de lockdowns foram positivas ou negativas, por conta de não termos muitos dados para comparação, praticamente todos os países do mundo adotaram a mesma estratégia de contenção da Covid-19. 

É uma doença que está ativa em nosso planeta e provavelmente não deixará de existir assim como as gripes mais tradicionais. A grande notícia de todo esse tempo é que felizmente por conta da tecnologia atual podemos ter acesso de forma rápida as vacinas, que com toda certeza contribuíram para uma redução do poder da doença. Hoje já começamos a ver novas opções de tratamento além das vacinas, vemos novos remédios sendo homologados e um melhor entendimento de como ela atua e se prolifera pelo planeta. 

Ainda temos uma quantidade significativa de novos casos, aparentemente com um índice de mortalidade muito mais controlado, por conta de hospitais e população mais preparados.

Acredito que teremos por um bom tempo um grande número de casos pelo mundo, mas com menos restrições e mais acesso a soluções para cura.
